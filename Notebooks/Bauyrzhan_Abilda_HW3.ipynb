```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bootstrap

pd.set_option("display.float_format", "{:.4f}".format)
sns.set_style("whitegrid")
```

```python
# Defining the correct dataset path relative to the Notebooks folder
data_path = "../Data/AReM/"

# Listing all activity folders to check if the data exists
activity_folders = [folder for folder in os.listdir(data_path) if not folder.startswith(".")]
print("Activity folders:", activity_folders)
```

```python
# Initializing lists to store training and test data
train_data = []
test_data = []

def load_data():
    """
    I loaded the dataset, split it into training and test sets, 
    and stored them in separate lists. I also handled errors 
    during CSV parsing by skipping problematic rows.
    """
    for activity in activity_folders:
        folder_path = os.path.join(data_path, activity)
        files = sorted(glob.glob(os.path.join(folder_path, "*.csv")))

        print(f"Activity: {activity}, Files Found: {len(files)}")

        if len(files) == 0:
            print(f"Warning: No CSV files found in {folder_path}")
            continue  # Skip this activity if no files are found

        # Splitting data based on assignment instructions
        if "bending1" in activity or "bending2" in activity:
            test_files, train_files = files[:2], files[2:]
        else:
            test_files, train_files = files[:3], files[3:]

        # Reading and appending files to respective lists
        for file in train_files:
            try:
                df = pd.read_csv(file, skiprows=5, on_bad_lines="skip")  # Skipping bad rows
                if not df.empty:
                    train_data.append((df, activity))
            except Exception as e:
                print(f"Error reading {file}: {e}")

        for file in test_files:
            try:
                df = pd.read_csv(file, skiprows=5, on_bad_lines="skip")
                if not df.empty:
                    test_data.append((df, activity))
            except Exception as e:
                print(f"Error reading {file}: {e}")

# Loading the data
load_data()

# Displaying the number of instances in each set
print(f"Train instances: {len(train_data)}")
print(f"Test instances: {len(test_data)}")
```

```python
# Defining the function to extract statistical features from each instance
def extract_features(data):
    """
    I extracted statistical features from each time series in the dataset.
    Before computing statistics, I converted all values to numeric 
    to prevent errors caused by non-numeric data.
    """
    if len(data) == 0:
        print("Warning: No data available for feature extraction!")
        return pd.DataFrame(), []  

    features, labels = [], []

    for df, activity in data:
        row = []
        
        # Ensure all columns are numeric
        df = df.apply(pd.to_numeric, errors="coerce")
        
        for col in df.columns:
            series = df[col].dropna().values  
            
            if len(series) > 0:  
                row.extend([
                    np.min(series), np.max(series), np.mean(series), 
                    np.median(series), np.std(series), 
                    np.percentile(series, 25), np.percentile(series, 75)
                ])
            else:  
                row.extend([np.nan] * 7)  

        features.append(row)
        labels.append(activity)

    return pd.DataFrame(features), labels

# Extracting features for training and test sets
train_features, train_labels = extract_features(train_data)
test_features, test_labels = extract_features(test_data)

# Ensure extraction was successful
if train_features.empty or test_features.empty:
    print("Warning: Feature extraction returned empty results! Check your dataset.")

# Detecting the number of time series in the dataset dynamically
num_time_series = train_features.shape[1] // 7  
print(f"Detected {num_time_series} time series per instance")
```

```python
# Dynamically adjusting column naming to match detected time series
if not train_features.empty:
    feature_names = [f"{stat}{i+1}" for stat in ["min", "max", "mean", "median", "std", "Q1_", "Q3_"] for i in range(num_time_series)]

    train_features.columns = feature_names
    test_features.columns = feature_names

    print(f"Feature extraction completed with {len(feature_names)} features per instance.")
```

```python
# Checking the standard deviation for each extracted feature
std_values = train_features.std()

# Identifying features with zero variance
zero_variance_features = std_values[std_values == 0].index.tolist()

if zero_variance_features:
    print("Removing features with zero variance:")
    print(zero_variance_features)
else:
    print("No features with zero variance detected.")

# Checking the first few rows of extracted features
print("\nInspecting the first few rows of extracted features:")
display(train_features.head())

# Removing features with zero variance
train_features_cleaned = train_features.loc[:, (std_values != 0)]
test_features_cleaned = test_features.loc[:, (std_values != 0)]

print(f"\nUpdating dataset: {train_features_cleaned.shape[1]} features remaining (original: {train_features.shape[1]})")

# Checking the distribution of remaining features
print("\nVerifying standard deviation of cleaned features:")
display(train_features_cleaned.std())

# Checking for missing values in the dataset
missing_values = train_features_cleaned.isnull().sum().sum()

if missing_values > 0:
    print(f"Detecting {missing_values} missing values. Handling may be required before classification.")
else:
    print("No missing values detected in the dataset.")

# Updating the dataset with the cleaned version
train_features = train_features_cleaned
test_features = test_features_cleaned

print("\nFinal check complete. The dataset is now cleaned and ready for classification.")
```

```python
# Checking for missing values before handling
missing_values = train_features_cleaned.isnull().sum().sum()

if missing_values > 0:
    print(f"Handling {missing_values} missing values...")

    # Choosing the best method to handle missing values
    train_features_cleaned = train_features_cleaned.fillna(train_features_cleaned.mean())
    test_features_cleaned = test_features_cleaned.fillna(test_features_cleaned.mean())

    print("Missing values filled using mean of each feature.")

# Final check
missing_values_after = train_features_cleaned.isnull().sum().sum()
if missing_values_after == 0:
    print("Final check complete: No missing values remain.")
else:
    print(f"Warning: {missing_values_after} missing values still exist.")

# Updating the dataset with the cleaned version
train_features = train_features_cleaned
test_features = test_features_cleaned

print("\nDataset is now fully cleaned and ready for classification.")
```

```python
# Computing standard deviation for each extracted feature
std_values = train_features.std()
std_values
```

```python
# Defining a function to compute bootstrapped confidence intervals for standard deviation
def bootstrap_ci(data, num_resamples=1000, confidence_level=0.90):
    """
    I used bootstrapping to estimate the 90% confidence intervals for 
    the standard deviation of each feature. This helped me understand 
    the variability of extracted features.
    """
    results = {}
    for feature in data.columns:
        resampled = bootstrap((data[feature].values,), np.std, confidence_level=confidence_level)
        results[feature] = resampled.confidence_interval
    return results

# Computing bootstrap confidence intervals for standard deviation
bootstrap_results = bootstrap_ci(train_features)

# Storing the results in a DataFrame
bootstrap_df = pd.DataFrame.from_dict(bootstrap_results, orient="index", columns=["Lower CI", "Upper CI"])
bootstrap_df
```

```python
# Selecting the three most important features based on standard deviation
top_features = std_values.nlargest(3).index.tolist()
print("Selected Features:", top_features)

# Plotting distributions of the selected features
plt.figure(figsize=(12, 5))
for feature in top_features:
    sns.kdeplot(train_features[feature], label=feature, fill=True)
plt.legend()
plt.title("Distribution of Selected Features")
plt.show()
```
